```python
# how merge 2 or more dataframe
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,on="id")
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,on="id",how="left")
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,how="right")
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,how="outer",indicator=True)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1.1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "class":["a","b","c","d"]})
pd.merge(df1,df2,on="id",suffixes=("x","y"))
```

```python
# how to combining data
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([5,6,7,8,9])
pd.concat([sr1,sr2],ignore_index=True)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "sn":["a","b","c","d"]})
pd.concat([df1,df2],ignore_index=True)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "sn1":["a","b","c","d"]})
pd.concat([df1,df2],sort=True)
```

```python
# how to join data
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id1":[1,2,3,4],
                "sn1":["a","b","c","d"]})
df1.join(df2)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id1":[1,2,3,4],
                "sn1":["a","b","c","d"]})
df1.join(df2,how="left")
```

```python
# how to append data
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "sn":["a","b","c","d"]})
df1.append(df2)
```

```python
df1=pd.DataFrame({"id":[1,2,3,4],
                 "sn":[5,6,7,8]})
df2=pd.DataFrame({"id":[1,2,3,4],
                "sn":["a","b","c","d"]})
df1.append(df2,ignore_index=True)
```

```python
# when both df name is same
df1.append(df2,sort=False)
```
